```python
%tensorflow_version 2.x
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
tf.__version__
```

## Importao dos dados

```python
base_de_dados = pd.read_csv('/content/drive/MyDrive/IA/Base_Treinamento_Supervisionado.csv', encoding="ISO-8859-1",sep=";")


# Substituir vrgulas por pontos em todas as colunas
base_de_dados = base_de_dados.replace(',', '.', regex=True)

municipios_criticos = {}
for i,j in zip(base_de_dados['IFGF'], base_de_dados['Municpio']):
  if(float(i)<0.4):
    municipios_criticos[j] = i

base_de_dados
```

```python
base_de_dados.info()
```

```python
base_de_dados.isnull().sum()
```

existem 45 ocorrencias de <code>Percentual das receitas oriundas de fontes externas (%)</code> e 18 <code>Mortalidade Infantil</code> no informadas

## Tratamento dos dados

```python
base_de_dados = base_de_dados.drop('Municpio', axis=1)

# Converter todas as colunas para tipo float
base_de_dados = base_de_dados.astype(float)
```

```python
media_perc = base_de_dados['Percentual das receitas oriundas de fontes externas (%)'].median()
media_mi = base_de_dados['Mortalidade Infantil'].median()

values = {'Percentual das receitas oriundas de fontes externas (%)': media_perc, 'Mortalidade Infantil': media_mi}

base_de_dados = base_de_dados.fillna(value=values)

base_de_dados
```

```python
result = []

for i in base_de_dados['IFGF']:
  if i<0.4:
    result.append(1)
  else:
    result.append(0)

base_de_dados['IFGF'] = result
```

```python
base_de_dados
```

## Transformao dos dados

```python
X = base_de_dados.iloc[:, 1:17].values

X
```

```python
y = base_de_dados.iloc[:, 0].values

y
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

O modelo sequencial  uma pilha linear de camadas, onde as camadas so adicionadas sequencialmente.


## Criao da rede neural

```python
classifier = tf.keras.models.Sequential()
classifier.add(tf.keras.layers.Dense(units=400, activation='relu', input_shape=(16, )))
classifier.add(tf.keras.layers.Dropout(0.2))
classifier.add(tf.keras.layers.Dense(units=500, activation='relu'))
classifier.add(tf.keras.layers.Dropout(0.1))
classifier.add(tf.keras.layers.Dense(units=500, activation='relu'))
classifier.add(tf.keras.layers.Dropout(0.2))
classifier.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

Adicionamos uma camada densa (fully connected)  rede neural com 400 unidades (neurnios). A funo de ativao utilizada  a funo ReLU (Rectified Linear Unit). A camada possui input_shape=(16, ), o que indica que a entrada tem 16 caractersticas.


```python
classifier.summary()
```

```python
classifier.compile(optimizer='Adam', loss='binary_crossentropy', metrics = ['accuracy'])
```

Estamos usando o otimizador Adam, que  uma combinao do algoritmo RMSProp com o algoritmo de momento.

## Treinamento da rede

```python
epochs_hist = classifier.fit(X_train, y_train, epochs = 50)
```

- X_train:  o conjunto de dados de treinamento contendo as caractersticas (features) usadas para treinar o modelo.

- y_train:  o conjunto de dados de treinamento contendo as classes ou rtulos correspondentes s caractersticas em X_train.

- epochs = 50: O parmetro epochs define o nmero de vezes que o conjunto de treinamento ser percorrido durante o treinamento do modelo


```python
epochs = epochs_hist.epoch
accuracy = epochs_hist.history['accuracy']
```

# Data Visualization!

```python
plt.plot(epochs, accuracy, marker='o')  # Grfico de linha com marcadores circulares
plt.xlabel('pocas')
plt.ylabel('Acurcia')
plt.title('Grfico de Classificao de Acurcia')
plt.grid(True)  # Adiciona uma grade ao grfico
plt.show()  # Mostra o grfico
```

## Validao com dados reais

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred = (y_pred > 0.5)

y_pred
```

Lembre-se de que as previses retornadas pelo mtodo predict() sero valores contnuos entre 0 e 1, pois estamos usando a funo de ativao sigmoid na ltima camada da rede neural. Se voc deseja obter previses binrias (0 ou 1), voc pode aplicar um limiar (threshold) s previses, por exemplo, usando y_pred = (y_pred > 0.5).astype(int). Isso ir converter os valores de y_pred em 0 ou 1 com base em um limiar de 0.5.

```python
y_pred = np.argmax(y_pred, axis=1)
```

```python
epochs = epochs_hist.epoch
train_accuracy = epochs_hist.history['accuracy']
val_accuracy = y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_train, y_pred)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Classe Predita')
plt.ylabel('Classe Verdadeira')
plt.title('Matriz de Confuso')
plt.show()
```
